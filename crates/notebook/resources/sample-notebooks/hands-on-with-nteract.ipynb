## 1. Basic markdown cells

Markdown cells are the narrative layer of a notebook.
Use them to explain what a code cell does, leave instructions for collaborators, and keep results interpretable.

Try editing this cell:
- change the heading
- add a bullet
- add `inline code`
- add a short note about what you want to explore next

````python
from datetime import datetime

numbers = [3, 5, 8, 13]
average = sum(numbers) / len(numbers)

print("Hello from nteract.")
print("Opened at:", datetime.now().isoformat(timespec="seconds"))
print("Numbers:", numbers)
print("Average:", round(average, 2))
````

## 2. Notebook environments and dependencies

This sample notebook is saved with two Python dependencies:
- `pandas`
- `matplotlib`

Run the next few cells to confirm those packages are available and to generate table and chart outputs.

````python
import pandas as pd
import matplotlib.pyplot as plt

steps = pd.DataFrame(
    [
        {"step": "Write markdown", "minutes": 3},
        {"step": "Run Python", "minutes": 5},
        {"step": "Check outputs", "minutes": 4},
        {"step": "Edit with MCP", "minutes": 6},
    ]
)
steps["minutes_per_step"] = steps["minutes"] / steps["minutes"].sum()
steps
````

````python
ax = steps.plot(
    kind="bar",
    x="step",
    y="minutes",
    legend=False,
    color=["#0f766e", "#2563eb", "#9333ea", "#ea580c"],
    figsize=(8, 4),
)
ax.set_title("A tiny notebook workflow")
ax.set_ylabel("Minutes")
plt.tight_layout()
plt.show()
````

## 3. Connect nteract MCP to your code agent

Use the published MCP server for your agent client. The commands below install the production `nteract` MCP entrypoint via `uvx`, rather than this repo's local development setup.

### Claude Code
```bash
claude mcp add nteract -- uvx nteract
```

### Codex CLI
```bash
codex mcp add nteract -- uvx nteract
```

Once your agent can see the nteract MCP server, it can use tools like:
- `open_notebook`
- `get_all_cells`
- `set_cell`
- `create_cell`
- `execute_cell`
- `save_notebook`

That lets your agent inspect this notebook, change cells, run code, and save the result back to disk.
